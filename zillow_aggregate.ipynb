```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import sklearn
import seaborn as sns
import scipy

# This is importing a "future" python version 3 print function.
from __future__ import print_function
from __future__ import division
%matplotlib inline
sns.set()
sns.set_context('paper', font_scale = 1.5)
sns.set_style('ticks')
```

```python
Bed1 = pd.read_csv('Neighborhood_MedianRentalPrice_1Bedroom_fits.csv')
Bed2 = pd.read_csv('Neighborhood_MedianRentalPrice_2Bedroom_fits.csv')
Bed3 = pd.read_csv('Neighborhood_MedianRentalPrice_3Bedroom_fits.csv')
```

```python
dd = Bed1.merge(Bed2, on = 'Name', how = 'inner')
d = dd.merge(Bed3, on = 'Name', how = 'inner')
d.drop_duplicates(['Name'], keep = 'first',inplace = True)
```

```python
plt.scatter(d['price_trend_1Bedroom'], d['price_trend_2Bedroom'],\
            cmap = 'jet', c = d['price_trend_3Bedroom'])
plt.colorbar()
```
